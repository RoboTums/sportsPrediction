# Using Sportsreference API to create a consistent estimation algorithm for the number of fantasy points each player in each position in each game has made

## We need to gather:
- Player data for each game.
- player data relating to their team.


```python
from sportsreference.nfl.boxscore import Boxscores, Boxscore
from sportsreference.nfl.roster import Player
from sportsreference.nfl.teams import Teams, Team
from sportsreference.nfl.schedule import Schedule
import pandas as pd
#games = Boxscores(1,2017)
#games.games
```

## gather all players on each team for each year from 2017-2020.
These API requests take some time. 

```python
player_collector = []
team_collector = []
game_collector = []
try:
    for team in Teams():
        roster = team.roster  # Gets each team's roster
        team_collector.append(team.dataframe)
        schedule =  team.schedule
        team_name = team_collector[-1].index[0]
        game_data_df = schedule.dataframe_extended
        game_data_df['team_name'] = team_name
        game_collector.append(game_data_df)
        for player in roster.players:
            if type(player) is not None:
                player_collector.append(player.dataframe)
        
except AttributeError:
    print('time to patch this api')
```

```python
for player in player_collector:
    player_id = player.player_id[0]
    player_name = player.name[0]
    player.to_csv(f'../data/sports_ref_data/player_data/{player_id}_{player_name}.csv')
```
